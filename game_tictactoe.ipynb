```python
import math

X = "X"
O = "O"
EMPTY = None

def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    x = sum(row.count(X) for row in board)
    o = sum(row.count(O) for row in board)
    if x == 0 and o == 0:
        return X
    elif x > o:
        return O
    else:
        return X

def actions(board):
    AllPossibleMoves = set()
    for r in range(3):
        for c in range(3):
            if board[r][c] == EMPTY:
                AllPossibleMoves.add((r, c))
    return AllPossibleMoves

def result(board, action):
    if board[action[0]][action[1]] != EMPTY:
        raise ValueError("Invalid Move: The given cell is not EMPTY.\n")
    deepCopy = [row[:] for row in board]
    move = player(board)
    deepCopy[action[0]][action[1]] = move
    return deepCopy

def winner(board):
    for i in range(3):
        if board[i].count(X) == 3:
            return X
        elif board[i].count(O) == 3:
            return O

    for col in range(3):
        column = [board[row][col] for row in range(3)]
        if column.count(X) == 3:
            return X
        elif column.count(O) == 3:
            return O

    xdp = 0
    odp = 0
    xds = 0
    ods = 0
    n = 3

    for i in range(3):
        if board[i][i] == X:
            xdp += 1
        elif board[i][i] == O:
            odp += 1
        if board[i][n - i - 1] == X:
            xds += 1
        elif board[i][n - i - 1] == O:
            ods += 1

    if xdp == 3 or xds == 3:
        return X
    if odp == 3 or ods == 3:
        return O
    return None

def terminal(board):
    isBoardFull = True
    for row in board:
        if EMPTY in row:
            isBoardFull = False
            break
    if winner(board) is not None or isBoardFull == True:
        return True
    else:
        return False

def utility(board):
    value = winner(board)
    if value == X:
        return 1
    elif value == O:
        return -1
    else:
        return 0

def minimax(board):
    if terminal(board):
        return None

    currentPlayer = player(board)
    possibleMoves = actions(board)

    bestScore = -math.inf if currentPlayer == X else math.inf
    bestMove = None

    for move in possibleMoves:
        newBoard = result(board, move)

        def minimax_value(board):
            if terminal(board):
                return utility(board)
            current = player(board)
            if current == X:
                best = -math.inf
                for action in actions(board):
                    val = minimax_value(result(board, action))
                    best = max(best, val)
                return best
            else:
                best = math.inf
                for action in actions(board):
                    val = minimax_value(result(board, action))
                    best = min(best, val)
                return best

        value = minimax_value(newBoard)

        if currentPlayer == X:
            if value > bestScore:
                bestScore = value
                bestMove = move
        else:
            if value < bestScore:
                bestScore = value
                bestMove = move

    return bestMove
```
